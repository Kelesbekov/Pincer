```python
import os
import csv
from collections import defaultdict
import matplotlib.pyplot as plt

# Get the current working directory
current_directory = os.getcwd()

print("Current working directory:", current_directory)

all_items = os.listdir(current_directory)

prefix_to_match = 'Participant-116'
matching_folders = [item for item in all_items \
                    if item.startswith(prefix_to_match)]


for file_path in matching_folders:
    columns_to_keep = ['Pinch Aperture']
    data = []
    print(file_path)
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        data = [row[col] for col in columns_to_keep for row in reader]
        for row in reader:
            data.append(row)
        print(len(data))
        print(data[0])
        plt.figure(figsize=(30,6))
        data = [float(item)*1000 for item in data]
        plt.plot(data)
```
